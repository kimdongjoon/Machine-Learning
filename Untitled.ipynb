```python
#Alice in Monderland
text = open('./wordCloudData/wc1.alice.txt').read()
alice_mask = np.array(Image.open('./wordCloudData/wc2.alice_mask.png'))

stopwords = set(STOPWORDS)
stopwords.add("said")

text_ = [word for word in text.split() if word.lower() not in stopwords]
text = ' '.join(str(elem) for elem in text_)

img = alice_mask
image_array = np.array(img)
```

```python
text_
```

```python
wc = WordCloud(background_color = 'white'
               , width = 2000
               , height = 2000
               , max_font_size = 150
               , scale = 2.0
               , max_words = 200
               , mask = alice_mask
               , stopwords = text
               , contour_color = 'steelblue'
              )
gen = wc.generate(text)
gen.words_

plt.figure(figsize=(12,22))
plt.imshow(gen, interpolation= 'bilinear')
plt.axis('off')
plt.show()
```
